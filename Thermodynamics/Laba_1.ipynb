```python
import re
```

```python
pattern = re.compile(r'[^'' .,*+-'']\d{4,}[^ ]')

s="234533 человека выпили 12,5 л пива 12312312 -> человека выпили 12,5л пива"

def replacer(match):
    return f'{int(match.group()):,}'.replace(',', ' ')

    
re.sub(pattern, replacer, s)
```
