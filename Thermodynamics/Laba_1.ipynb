```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
```

```python
df = pd.read_csv('Data_1.csv')
```

```python
x = df.time.values[3:]*60
y = df.preasure.values[3:]*0.01
x = x + 20
```

```python
plt.plot(x, y, "o-")
plt.title('Зависимость давления от времени откачки в камере (с ПМТ-2)')
plt.xlabel("Время(с)")
plt.ylabel("Давление в камере(мбар)")
```

```python
plt.loglog(x, y, "o-")
plt.title('Зависимость давления от времени откачки в камере (с ПМТ-2)')
plt.xlabel("Время(с)")
plt.ylabel("Давление в камере(мбар)")
```

```python
y_2 = -(np.log(y/y[0]))/x
```

```python
plt.plot(y, y_2, "o-")
plt.title("Зависимость скорости откачка от давления в камере (ПМТ-2)")
plt.xlabel("Давление (мбар)")
plt.xticks(rotation=45)
plt.ylabel("Быстрота откачки")
```

```python
plt.plot(x, y_2, "o-")
plt.title("Зависимость скорости откачка от времени в камере (ПМТ-2)")
plt.xlabel("Время (с)")
plt.ylabel("Быстрота откачки")
```
